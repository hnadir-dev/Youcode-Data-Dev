```python
import pandas as pd
```

```python
df_data_without_watchers_outliers = pd.read_csv('data/data_cleaned/data_without_watchers_outliers_2023_05.csv')
df_watchers_outlires = pd.read_csv('data/data_cleaned/watchers_outliers_2023_05.csv')

contributors = pd.read_csv('data/data_cleaned/repo_contributors_05.csv')

data_c = pd.concat([df_data_without_watchers_outliers,df_watchers_outlires])

#data_c['full_name'].drop_duplicates(data_c['full_name'].index,inplace=True)
data_c.drop(columns=['Unnamed: 0.1','Unnamed: 0'],inplace=True)

#data_c['license'].replace('No License',{'key': 0,'name': 'No License','spdx_id': '','url': '','node_id': ''},inplace=True)

data_c.drop_duplicates(inplace=True)
#data_c.info()
```

```python
#Replace No License to dict
data_c['license'].replace('No License',str(dict(key= '',name= 'No License',spdx_id= '',url= '',node_id= '')),inplace=True)
```

<h4>Technologies</h4>

```python
technoligies = pd.DataFrame(list(data_c['language'].unique())).rename(columns={0:'name'})
techs = []
for index,tech in technoligies.iterrows():
    techs.append(dict(id=index+1,name=tech['name']))

technoligies = pd.DataFrame(techs)
technoligies
```

<h4>Owners</h4>

```python
owners = pd.DataFrame(list(data_c['type'].unique())).rename(columns={0:'name'})
ors = []
for index,otype in owners.iterrows():
    ors.append(dict(id=index+1,name=otype['name']))

owners = pd.DataFrame(ors)
owners
```

<h4>License</h4>

```python
import ast

#table license
licenses = pd.DataFrame(list(data_c['license'].unique())).rename(columns={0:'name'})
lics = []
for index,l in licenses.iterrows():
    lics.append(ast.literal_eval(l['name']))


licenses = pd.DataFrame(lics)
licenses.insert(0,"id", list(range(1,len(lics)+1)),True)
licenses
```

<h4>Contributors</h4>

```python
cs = pd.DataFrame(contributors['name'].unique()).rename(columns={0:'name'})

cs.insert(0,'id',range(1,len(cs)+1),True)

cs
```

<h4>Repositories</h4>

```python
repositories = pd.DataFrame(data_c['full_name'].unique()).rename(columns={0:'full_name'})

repositories.insert(0,'id',range(1,len(repositories)+1),True)
repositories
```

```python
dataToInserted = data_c

#--Replace technologies
for index,tech in technoligies.iterrows():
    dataToInserted['language'].replace(tech['name'],tech['id'],inplace=True)


#--Replace owners
for index,tech in owners.iterrows():
    dataToInserted['type'].replace(tech['name'],tech['id'],inplace=True)

dataToInserted.head(2)
```

```python
#--Replace license
for index,l in licenses.iterrows():
    #print(str(dict(l[['key','name','spdx_id','url','node_id']])))
    dataToInserted['license'].replace(str(dict(l[['key','name','spdx_id','url','node_id']])),l['id'],inplace=True)

#dataToInserted[dataToInserted['license'] == str(dict(licenses[['key','name','spdx_id','url','node_id']].head(1)))]

#dict(licenses[['key','name','spdx_id','url','node_id']].head(1))
dataToInserted.head(2)
```

```python
# add id to each repos
#add columns ID
dataToInserted['id'] = dataToInserted['full_name']

for index, r in repositories.iterrows():
    dataToInserted['id'].replace(r['full_name'],r['id'],inplace=True)
```
